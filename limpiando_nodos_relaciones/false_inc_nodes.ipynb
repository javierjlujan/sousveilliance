```python
import pandas as pd
```

```python
df_relaciones = pd.read_csv('../datos/relaciones.csv', names=['numero_correlativo', 'dni', 'tipo'])
df_relaciones.head()
```

```python
df_nodos_empresas = pd.read_csv('../datos/nodos_empresas.csv', names=['nodo', 'dni', 'persona_fisica', 'tipo'])
df_nodos_empresas.head()
```

```python
numeros_relaciones = set(df_relaciones['numero_correlativo'].unique())
numeros_nodos = set(df_nodos_empresas['dni'].unique())

no_existen = numeros_relaciones.difference(numeros_nodos)
no_existen
```

```python
def filtrando(ele):
    if ele in no_existen:
        return False
    else:
        return True
    
df_relaciones['no_existe'] = df_relaciones['numero_correlativo'].apply(filtrando)
df_relaciones.head()
```

```python
df_relaciones[df_relaciones['no_existe'] == False]
```

```python
df_limpio = df_relaciones[df_relaciones['no_existe'] == True]
df_limpio = df_limpio.drop('no_existe', axis=1)
df_limpio.head()
```
