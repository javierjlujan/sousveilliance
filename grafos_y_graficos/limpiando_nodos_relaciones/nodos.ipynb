```python
import pandas as pd
```

```python
df_autoridades = pd.read_csv('../datos/igj-autoridades.csv')

def remove_spaces(ele):
    ele = ele.replace('.', '')
    return ''.join(ele.split())

df_autoridades['numero_documento'] = df_autoridades['numero_documento'].astype(str)
df_autoridades['numero_documento'] = df_autoridades['numero_documento'].apply(remove_spaces)
df_autoridades.head()
```

```python
df_nodos = pd.DataFrame()
df_nodos['nodos'] = df_autoridades.drop_duplicates('numero_documento', keep='first')['apellido_nombre']
df_nodos['dni'] = df_autoridades.drop_duplicates('numero_documento', keep='first')['numero_documento']
df_nodos['persona_fisica'] = [True for _ in range(df_nodos.shape[0])]
df_nodos['tipo'] = df_autoridades.drop_duplicates('numero_documento', keep='first')['tipo_documento']

df_nodos
```

```python
df_nodos.to_csv('../datos/nodos_personas.csv', index=False, header=False)
```

```python
df_entidades = pd.read_csv('../datos/igj-entidades.csv')
df_entidades.head()
```

```python
df_nodos = pd.DataFrame()

df_nodos['nodos'] = df_entidades.drop_duplicates('numero_correlativo', keep='first')['razon_social']
df_nodos['dni'] = df_entidades.drop_duplicates('numero_correlativo', keep='first')['numero_correlativo']
df_nodos['persona_fisica'] = [False for _ in range(df_nodos.shape[0])]
df_nodos['tipo'] = df_entidades.drop_duplicates('numero_correlativo', keep='first')['tipo_societario']

df_nodos.head()
```

```python
df_nodos.to_csv('../datos/nodos_empresas.csv', index=False, header=False)
```
